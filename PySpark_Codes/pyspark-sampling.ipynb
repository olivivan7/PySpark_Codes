```python
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .master("local[1]") \
    .appName("SparkByExamples.com") \
    .getOrCreate()
```

```python
df=spark.range(100)
```


sample() 
<br>
print(df.sample(0.06).collect())<br>
print(df.sample(0.1,123).collect())<br>
print(df.sample(0.1,123).collect())<br>
print(df.sample(0.1,456).collect())<br>
print("withReplacement Examples")<br>
print(df.sample(True,0.3,123).collect())<br>
print(df.sample(0.3,123).collect())<br>
ampleBy() 


```python
print("sampleBy Examples")
df2=df.select((df.id % 3).alias("key"))
print(df2.sampleBy("key", {0: 0.1, 1: 0.2},0).collect())
```

```python
print("RDD Examples")
'''RDD'''
rdd = spark.sparkContext.range(0,100)
print(rdd.sample(False,0.1,0).collect())
print(rdd.sample(True,0.3,123).collect())
```
